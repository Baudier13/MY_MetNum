## Parte I

El problema consiste en encontrar una raíz de una función dada de acuerdo a distintos
métodos y estudiar aspectos de estos.

1. Método de Newton.

2. Método extendido de Newton.

El método extendido de Newton fue ideado para encontrar raíces 
múltiples de una función $f(x)$ generando una secuencia $\{x_n\}_0^{\infty}$ dada por
$$
x_{n+1} = x_{n} - \frac{f(x_n)\,f^{\prime}(x_n)}{(f^{\prime}(x_n))^2 - f(x_n)\, f^{\prime\prime}(x_n)}
$$

**I.A)** Para cada uno de los métodos mencionados escriba una función que permita obtener una aproximación a la raíz de una función arbitraria, $f(x)$. 


Estas funciones deberán:

1. Contener como argumentos: 

    a. la función a la que deseamos encontrar la raíz y sus derivadas según corresponda,
    
    b. una aproximación inicial $x_0$,
    
    d. la tolerancia en $x$, tol$_x$ (este argumento debe ser opcional y tener un valor prefijado por defecto),
    
    e. la tolerancia en $f(x)$, tol$_f$ (idem anterior),
    
    f. y el número máximo de iteraciones, $N_{\max}$ (también prefijado por defecto).

Cualquiera sea el método, el algoritmo debe deternerse cuando el error relativo en $x$, aquí denotado por $E_{r,x}$, sea menor que tol$_{x}$ y el error absoluto en $f(x)$, aquí denotado por $E_{a,f}$, sea menor a tol$_{f}$, o cuando el número de iteracciones alcance el valor $N_{\max}$. 
En otras palabras, la iteración debe detenerse cuando se satisfaga que:

\begin{equation*}
E_{r,x} = \frac{|x_n - x_{n-1}|}{|x_n|} < \mbox{tol}_x, 
\;\;\;\;\; \mbox{y} \;\;\;\;\;
E_{a,f}=|f(x_n)| < \mbox{tol}_f , 
\end{equation*}
ó cuando se llegue al número máximo de iteraciones $n=N_\text{max}$.

Además, la función debe:

2. usando `println`, imprimir los siguientes valores finales: 

    a. en número $N$ de iteraciones realizadas, 

    b. el valor de la aproximación final $x_N$, 

    c. el valor de la función en la aproximación final $f(x_N)$,
    
    d. el error relativo en $x$ final (aproximado), 

    e. el error absoluto en $f$ final (aproximado).

3. imprimir una señal de aviso si el método no convergió en $N_{\max}$ iteraciones,

4. y retornar vectores con los errores en cada paso de iteración: $E_{r,x}[n]$, $E_{a,f}[n]$.

```julia
# Escriba aquí (en un par de celdas) las funciones que implementan los metodos.

#Método Newton

function newton(f::Function,  # Defino las variables e inicializo las opcionales.
    der_f::Function,
    x_0::Float64;
    tol_x::Float64=10^(-10),
    tol_y::Float64=10^(-10),
    maxiter::Int64=100)
    
    
    p0 = x_0
    p1 = p0  # solo lo inicializo para que quede declarada la variable, pero sera reescrita en cada paso
    
    #Defino los vectores:
    errores_relativos = Vector(undef,maxiter)
    errores_absolutos = Vector(undef,maxiter)
    
    
    fp0 = f(p0)
    error_abs = abs(f(p1))
    error_rel = abs(p1-p0)/abs(p1) #error relativo
    i = 1
    while i < maxiter
        if der_f(x_0) == 0
            error("La derivada es igual a cero")
        end
       
        p1 = p0 - ((fp0)/(der_f(p0)))
        error_abs = abs(p1-p0)
        error_rel = abs(p1-p0)/abs(p1)
        
        
        p0 = p1
        fp0 = f(p0)
        
        errores_relativos[i] = error_rel
        errores_absolutos[i] = error_abs
        
        if (error_rel < tol_x) && (error_abs < tol_y)
            break 
        end
        i += 1
    end
    if (i >= maxiter) && (error_abs > tol_y)
        error("La función no convergió en el máximo de iteraciones solicitado")
    end
        
    absoluto_raiz = abs(fp0)
    println("iteraciones $i,  ",
        "la raiz es: $p0,  ",
        "valor de la función: $fp0,  ",
        "valor absoluto de la raiz $absoluto_raiz,  ",
        "error relativo $error_rel,  ",
        "error absoluto $error_abs,  ")
    
    return[errores_relativos[1:i],
        errores_absolutos[1:i]]
end
```

```julia
# Método Newton extendido
function newton_multiple(
    f::Function,
    df::Function,
    ddf::Function,    
    x_0::Float64;
    tol_x::Float64=10^(-10),
    tol_y::Float64=10^(-10),
    maxiter::Int64=100)
    
    p0 = x_0
    fp0 = f(p0)
    dfp0 = df(p0)
    ddfp0 = ddf(p0)
    p1 = p0 # solo lo inicializo para que quede declarada la variable, pero sera reescrita en cada paso
    
    #Defino los vectores:
    errores_relativos = Vector(undef,maxiter)
    errores_absolutos = Vector(undef,maxiter)
    fp1 = f(p1)
    error_abs = abs(f(p1))
    error_rel = abs(p1-p0)/abs(p1) #error relativo
    i = 1
    
    while i < maxiter
        if df(x_0) == 0
            error("La derivada es igual a cero")
        end
       
        p1 = p0 - ((fp0 * dfp0)/(dfp0^2 - fp0 * ddfp0))   # Paso iterativo
        fp1 = f(p1)
        error_abs = abs(fp1)
        error_rel = abs(p1-p0)/abs(p1)
        
        # vuelvo a calcular las variables y derivadas
        p0 = p1
        fp0 = f(p0)
        dfp0 = df(p0)
        ddfp0 = ddf(p0)
        
        errores_relativos[i] = error_rel
        errores_absolutos[i] = error_abs
        
        if (error_rel < tol_x) && (error_abs < tol_y)
            break 
        end
        
        i += 1 
    end
    
    if (i > maxiter) && (error_abs > tol_y)
        error("La función no convergió en el máximo de iteraciones solicitado")
    end
    
    absoluto_raiz = abs(fp0)
    println("iteraciones $i,  ",
        "la raiz es: $p0,  ",
        "valor de la función: $fp0,  ",
        "valor absoluto de la raiz $absoluto_raiz,  ",
        "error relativo $error_rel,  ",
        "error absoluto $error_abs")
    
    return[errores_relativos[1:i],
        errores_absolutos[1:i]]
end
    
```

```julia
function QuitarCeros(v::Array{Any,1})
    N = length(v)
    sinceros = Vector()
    for i ∈ 1:N
        if !(v[i]==0)
            push!(sinceros,v[i])
        end
    end
    return[sinceros]
end 
```

```julia
p3MultipleError = QuitarCeros(p3newtonMultiple[1])
```

```julia
scatter(p3newton[1],
    yaxis=:log,
    ylims = (10e-7, 1),
    label = "Newton",
    xlab = "Número iteraciones",
    ylab = "Valor del error relativo",
    title = "Comparación de errores de ambos métodos"
    )
scatter!(p3MultipleError, label = "Newton_múltiple") 
```

## Parte IV

**IV.A)** Utilizando una celda en formato *markdown*, agregue una explicación de la diferencia de comportamiento del error relativo $E_{r,x}$ al comparar lo observado en los gráficos de las secciones **II.C)** y **III.C)**.

Al poner escalas logarítmicas y ver que las curvas se asemejan a rectas sabemos que los errores tienen un comportamiento exponencial.

Para la primera función vemos que los errores tienen una diferencia mínima pero el metodo de newton múltiple es menor. Al no ser un punto de raices múltiples, ningun método tiene preferencia para encontrarlo, pero para el método de newton múltiple empleamos la derivada y la derivada segunda. Al utilizar ambas derivadas estamos dando mas información sobre la función en cada paso de la iteración y en cada paso es mejor definido el punto y el error es menor.

En la parte 3 ya la diferencia entre los métodos es sustancial. Para empezar podemos ver el numero de iteraciones: en newton necesitamos 21 iteraciones y en newton múltiple necesitamos solo 5. Al ser ambas rectas en escala logarítmica volvemos a obtener el comportamiento exponencial de los errores, lo cual esta bien. Pero el error decae mas rapidamente con el método de newton múltiple. Esto se debe a que estamos calculando el valor de una raiz múltiple de orden 2 y en este caso el método de newton no converge de forma cuadrática. Para estas situaciones es que definimos el método de newton múltiple en el que definimos una nueva funcion (en general):
$$ M(x) = \frac{f(x)}{f'(x)} $$
donde f(x) era la función a la cual queriamos calcular el cero de orden múltiple. Ahora por como la definimos sabemos que M(x) tiene un cero de orden 1 donde teniamos el cero de orden m en f(x), entonces si aplicamos el método de newton a M(x) obtenemos la convergencia cuadrática que buscabamos. El método de newton aplicado a M(x) nos da un paso iterativo que reemplazando nuevamente M(x) por su expresión (con la función f) obtenemos ahora el método de newton múltiple para la función f(x), que converge cuadráticamente.
Volviendo al ejemplo, tenemos una raiz múltiple y ambos métodos convergen, solo que newton múltiple converge de forma cuadrática, mientras que newton simple no, y por esto encuentra la raiz mas rapidamente.
