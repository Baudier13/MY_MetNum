# Problema 1
1. Para las funciones $f(x) = \ln (x+1)$ y $g(x) = \sqrt{x+1}$ y los puntos $x_0 = 0$, $x_1=0.6$ y $x_2=0.9$, construya analíticamente los polinomios interpolantes de Lagrange de grado 1 y 2 que aproximan la función en $x=0.45$.

2. Encuentre los errores absolutos y relativos correspondientes.

3. Grafique ambas funciones, sus polinomios interpolantes y la aproximación de Taylor de grado 2 (entorno a $x_0$) en el rango dado.

```julia
f(x) = log(x + 1)
g(x) = sqrt(x + 1)
x0 = 0
x1 = 0.6 
x2 = 0.9
xr = 0.45
```

```julia
L1(x)= ((x - x1)/(x0 - x1))*((x - x2)/(x0 - x2))
L2(x)= ((x - x0)/(x1 - x0))*((x - x2)/(x1 - x2))
L3(x)= ((x - x0)/(x2 - x0))*((x - x1)/(x2 - x1))
```

```julia
PLf(x)= f(x0)*L1(x) + f(x1)*L2(x) + f(x2)*L3(x)
PLg(x)= g(x0)*L1(x) + g(x1)*L2(x) + g(x2)*L3(x)
```

```julia
println(PLf(xr),"   ",PLg(xr))
```

```julia
eabsf = abs(f(xr)-PLf(xr))
erelf = abs(f(xr)-PLf(xr))/abs(f(xr))
eabsg = abs(g(xr)-PLg(xr))
erelg = abs(g(xr)-PLg(xr))/abs(g(xr))
```

```julia
println(eabsf,"  ",erelf,"  ",eabsg,"  ",erelg)
```

1. Escriba una función que evalúe el **polinomio interpolante de Lagrange** $PL$ en un punto $x$ con $x_0 < x < x_n$ siendo $(x_i,y_i)$ para $i=0,...,n$ los puntos a interpolar.
La función debe tomar como argumentos de entrada: *i)* el valor $x$, *ii)* un arreglo de valores $x_i$ y *iii)* un arreglo de valores $y_i$, y debe retornal el valor $PL(x)$.

2. Para cada función del **Problema 1**, realice un gráfico de $PL$ sobre $N=200$ puntos equidistantes en el intérvalo $[x_0,x_n]$ interpolando los puntos $(x_i,y_i)$. Incluya en curvas punteadas la función y en símbolos para los puntos interpolantes.

3. Gráfique la diferencia entre los polinomios y las funciones.

```julia
function Li(i::Int64,
    x::Float64,
    xvec::Vector{Float64}
    )::Float64
    L = 1.0
    xi = xvec[i]
    for j ∈ 1:i-1
        xj = xvec[j]
        L *= (x-xj)/(xi-xj)
    end
    for j ∈ i+1:length(xvec)
        xj = xvec[j]
        L *= (x-xj)/(xi-xj)
    end
    return L
end
```

```julia
function PL(x::Float64,
    xvec::Vector{Float64},
    yvec::Vector{Float64}
    )::Float64
    Pl = 0.0
    for i ∈ 1:length(xvec)
        Pl += yvec[i]*Li(i,x,xvec)
    end
    return Pl
end
```

```julia
f(x) = log(x + 1)
n = 200
puntos = [0.1 + 1 * (i-1)/(n-1) for i in 1:n]
valorfuncion = f.(puntos)
PLf(x) = PL(x,puntos,valorfuncion)
```

```julia
plot(puntos,PLf.(puntos),
    lw = 5,
    #xlims = (0.4,.05)
    )
plot!(f, lw=2)
```

```julia
g(x) = sqrt(x + 1)
n = 200
puntos = [0.1 + 1 * (i-1)/(n-1) for i in 1:n]
valorfuncion = g.(puntos)
PLg(x) = PL(x,puntos,valorfuncion)
```

```julia
plot(puntos,PLg.(puntos),
    lw = 5,
    #xlims = (0.4,.05)
    )
plot!(g, lw=2)
```

# Problema 3
Se desea aproximar $\cos(x)$ en el intervalo $[0,1]$ con un error absoluto menor a $1\times 10^{-7}$ para todo $x \in [0,1]$. 
1. Usando el teorema del error de la interpolación polinomial, estime el número de puntos de interpolación que son necesarios para conseguir el máximo error absoluto mencionado.
2. Grafique el error absoluto en el intervalo en cuestión para tres casos particulares de $\{x_i\}$: *i)* puntos equidistantes $x_i=i/n$, *ii)* puntos al azar y *iii)* puntos distribuidos heterogéneamente $x_i=1/i$.

#### Teorema
Como aproximamos por funciones lineales entre cada punto el termino del error tiene siempre la segunda derivada.
y como suponemos que todos los valores son xespaciados, tenemos x(j) < x < x (j+1)
con x(j) = j * h y x(j+1) = (j+1)h con h el ancho del intervalo
entonces la cota del error es:
$$ |\frac{f^{(2)}(\zeta(x))}{2}| ~|x-jh| ~|x-(j+1)h| $$
Donde voy a estar buscando el maximo en el intervalo para valuar la derivada y asi tener una cota.
Por otro lado, trabajando con el polinomio de los x, j y h tenemos que el maximo de ese polinomio esta siempre a mitad del intervalo en el punto h(j+1/2) y tenemos que:
$$ g(h(j+1/2)) = \frac{h^2}4 $$
entonces el error es acotado por:
$$ |f(x)-P(x)| <= |max(f^{(2)}(\zeta(x))|~~\frac{h^2}8 $$
Si ahora pedimos que ese error sea menor a un cierto valor, podemos despejar h y luego obtenemos que el numero de puntos que necesitamos es:
$$ N = int(1/h)+1 $$

```julia
# hacemos esto para el coseno, su derivada segunda es -cos(x) y en el intervalo [0,1] su máximo valor es 1
# y como el error que buscamos es 10e-7
h = sqrt(10e-7 * 8)
N = floor(Int64,1/h) + 1
```

i) Para puntos equidistantes el error depende de la cantidad de puntos, capaz se puede graficar la dependencia del error de cada intervalo con el numero de puntos que interpolamos, pero el error de cada intervalo es el mismo.

```julia
# Quiero un algoritmo para calcular las diferencias divididas para cada valor 
function dif_div(n::Int64,
        xvals::Vector{Float64},
        yvals::Vector{Float64}
        )
    i=1
    f = zeros(n,n)
    f[:,1]=yvals
    #for i ∈ 1:n
    #    f[i,1]=yvals[i]
    #end
    for i ∈ 1:n
        for j ∈ 2:i
            f[i,j] = (f[i,j-1]-f[i-1,j-1])/(xvals[i]-xvals[i-j+1])
        end
    end
    return f
end
# Esta función retorna f que es una matriz nxn donde la diagonal son las diferencias divididas necesarias.
# Entonces para formar el polinomio interpolante para una funcion g es:
# PNg(x) = f[1,1] + (x-x0)*f[2,2] + (x-x0)*(x-x1)*f[3,3]
```

```julia
function PN(x::Float64,
    n::Int64,
    xvals::Vector{Float64},
    yvals::Vector{Float64},
    )
    M = dif_div(n,xvals,yvals)
    P=M[1,1]
    N=1
    for i ∈ 1:(n-1)
        N=1
        for j ∈ 1:i
            N *= (x-xvals[j])
        end
        P += M[i+1,i+1]*N
    end
    return P
end
```

## Problema 2
**Error  de la interpolación polinomial para puntos equiespaciados.**

Considere el siguiente **teorema** dado en el teórico. 
Existe $\zeta$ dentro del intervalo de la interpolación tal que si la función es $n+1$ veces contínuamente diferenciable entonces, 

$$
f(x) - P_n(x) = \frac{f^{(n+1)}(\zeta)}{(n+1)!}\prod_{i=0}^{n} (x-x_i).
$$

Use el teorema para demostrar el siguiente **corolario**.

1. Sea $f(x)\, \varepsilon \, C_{[a,b]}^{(n+1)}$ tal que su derivada $n+1$ es acotada en $[a,b]$.
Es decir, $\exists M>0 : |f^{(n+1)}(x) |< M \;\forall \,x \,\varepsilon [a,b]$.

2. Sea $x_i=a + ih \;; i=0,\cdots,n$ donde $ h=(b-a)/n$. 

3. Sea $P_n(x)$ el polinomio interpolante a $f(x)$. 
Es decir, $P_n(x_i)=f(x_i)\;,i=0,\cdots ,n$. 

Entonces, $\forall\;x \,\varepsilon [a,b]$ se tiene
$$
\left| f(x) - P_n(x)\right | \leq \frac{M}{4 (n+1)}\;\left(\frac{b-a}{n}\right)^{n+1}.
$$
