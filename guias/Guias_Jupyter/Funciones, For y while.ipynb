Vamos a ver For, While y Funtions

$$
\sum_{i=1}^{200} 1+ \frac{1}{i^2}
$$

```julia
s = 0
for i ∈ 1:200
    s = s + 1/i^2
    println("i = $i, s = $s")
end    
```

```julia
i = 1
s = 0
while i <= 200
    s = s + 1/i^2
    println("i = $i, s = $s")
    i = i + 1
end   
```
